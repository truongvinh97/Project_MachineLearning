## Introduction


This data set contain information about 4804 movies collected from The Movies Database (TMDb), including user rating and revenue. This dataset is extract from kaggle: https://www.kaggle.com/tmdb/tmdb-movie-metadata. In this data set, I'll try to analize the data set to answer the questions:

a. What areas have the most influence on revenue?
b. How is a movie’s revenue and average score affected by its genre?
c. What influence does release date have on revenue?

## Describe

This dataset have some of the attribute:

    - budget: Budget of the movie
    - genres: Type of movie
    - homepage: Website for the movie
    - id: Index or id
    - original_language: Original language of the movie
    - original_title: Original name of the movie
    - popularity: Popularity of the movie
    - production_companies: Companies that have invested to the film
    - production_countries: The film's country
    - release_date: Date release  of the movie
    - revenue: Revenue by the movie
    - runtime: Film length or duration
    - status: Status of the movie
    - vote_average: Average Score of the movie 
    - vote_count: Number of vote

### What areas have the most influence on revenue?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import plotly.express as px
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy import stats

df_movie  = pd.read_csv('tmdb_5000_movies.csv')
df_movie.head()
test_1 = df_movie[['production_countries','revenue']]
display(test_1)
    
#data = pd.DataFrame(country_df)
#display(data)
```

### How is a movie’s revenue and average score affected by its genre?

```python
test_2 = pd.DataFrame(df_movie)
display(test_2[['genres','vote_average','revenue']])
```

### What influence does release date have on revenue?

```python
test_3 = pd.DataFrame(df_movie[['release_date','revenue']])
display(test_3)
```

```python
fig = px.scatter(test_3, x='release_date', y='revenue', range_x=['1930-01-01', '2017-12-31'],
                 title="Release date and Revenue")
fig.show()
```

```python
test_3_1=ols('revenue ~ release_date', data=test_3).fit()
print(sm.stats.anova_lm(test_3_1, typ=2))
```
